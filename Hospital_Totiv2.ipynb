##Parte 2: Geração de Insights comparativos dentro da base de dados do hospital



```python
import pandas as pd

df = pd.read_sql('''SELECT
  P.EmployeeID AS ProfessionalID,
  P.Name AS ProfessionalName,
  COUNT(Pr.Patient) AS PatientCount
FROM
  Physician P
LEFT JOIN
  Prescribes Pr ON P.EmployeeID = Pr.Physician
LEFT JOIN
  Undergoes U ON P.EmployeeID = U.Physician
GROUP BY
  P.EmployeeID, P.Name;''', conn)

print(df)

```

```python
import matplotlib.pyplot as plt



# Execute the query to retrieve the data
query = '''
SELECT
  P.EmployeeID AS ProfessionalID,
  P.Name AS ProfessionalName,
  COUNT(Pr.Patient) AS MedCount,
  COUNT(U.Patient) AS ProcedureCount
FROM
  Physician P
LEFT JOIN
  Prescribes Pr ON P.EmployeeID = Pr.Physician
LEFT JOIN
  Undergoes U ON P.EmployeeID = U.Physician
GROUP BY
  P.EmployeeID, P.Name;
'''
cursor.execute(query)
results = cursor.fetchall()
print(results)

df = pd.read_sql_query(query, conn)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a figure for prescriptions
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='ProfessionalName', y='MedCount', palette='Blues')
plt.title('Prescrições de Profissionais')
plt.xlabel('Nome')
plt.ylabel('Número de prescrições')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='ProfessionalName', y='ProcedureCount', palette='Reds')
plt.title('Procedimentos efetuados')
plt.xlabel('Nome')
plt.ylabel('Número de procedimentos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query2 = '''
SELECT
  P.EmployeeID AS ProfessionalID,
  P.Name AS ProfessionalName,
  AVG(Proc.Cost) AS AverageProcedureCost
FROM
  Physician P
JOIN
  Undergoes U ON P.EmployeeID = U.Physician
JOIN
  Procedure Proc ON U.Procedure = Proc.Code
GROUP BY
  P.EmployeeID, P.Name;
'''
cursor.execute(query)
results2 = cursor.fetchall()
print(results)

df2 = pd.read_sql_query(query2, conn)
print(df2)

plt.figure(figsize=(10, 6))
sns.barplot(data=df2, x='ProfessionalName', y='AverageProcedureCost', palette='Reds')
plt.title('Custo médio de procedimentos efetuados')
plt.xlabel('Nome')
plt.ylabel('Custo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query3 = '''
SELECT
  P.Code AS ProcedureCode,
  P.Name AS ProcedureName,
  P.Cost AS ProcedureCost,
  COUNT(U.Procedure) AS ProcedureCount
FROM
  Procedure P
LEFT JOIN
  Undergoes U ON P.Code = U.Procedure
GROUP BY
  P.Code, P.Name, P.Cost;
'''
cursor.execute(query)
results3 = cursor.fetchall()

df3 = pd.read_sql_query(query3, conn)

plt.figure(figsize=(10, 6))
sns.barplot(data=df3, x='ProcedureName', y='ProcedureCost', palette='Greens')
plt.title('Custo de procedimentos')
plt.xlabel('Nome')
plt.ylabel('Custo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query3 = '''
SELECT
  P.Code AS ProcedureCode,
  P.Name AS ProcedureName,
  P.Cost AS ProcedureCost,
  COUNT(U.Procedure) AS ProcedureCount
FROM
  Procedure P
LEFT JOIN
  Undergoes U ON P.Code = U.Procedure
GROUP BY
  P.Code, P.Name, P.Cost;
'''
cursor.execute(query)
results3 = cursor.fetchall()

df3 = pd.read_sql_query(query3, conn)

plt.figure(figsize=(10, 6))
sns.barplot(data=df3, x='ProcedureName', y='ProcedureCost', palette='Greens')
plt.title('Custo de procedimentos')
plt.xlabel('Nome')
plt.ylabel('Custo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query4 = '''
SELECT B.Floor, B.Code, AVG(ROUND((julianday(S.End) - julianday(S.Start)), 2)) AS AverageStayTime
FROM Stay AS S
JOIN Room AS R ON S.Room = R.Number
JOIN Block AS B ON R.BlockFloor = B.Floor AND R.BlockCode = B.Code
GROUP BY B.Floor, B.Code;

'''
cursor.execute(query4)
results4 = cursor.fetchall()

df4 = pd.read_sql_query(query4, conn)
print(df4)
plt.figure(figsize=(10, 6))
sns.barplot(data=df4, x='Code', y='AverageStayTime', palette='Greens')
plt.title('Media da estadia em cada bloco')
plt.xlabel('Bloco')
plt.ylabel('Media de estadia (em dias)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query5 = '''
SELECT Medication.Name, AVG(CAST(Prescribes.Dose AS DECIMAL)) AS AverageDose
FROM Medication
JOIN Prescribes ON Medication.Code = Prescribes.Medication
GROUP BY Medication.Name;
'''
cursor.execute(query5)
results5 = cursor.fetchall()

df5 = pd.read_sql_query(query5, conn)
print(df5)
plt.figure(figsize=(10, 6))
sns.barplot(data=df5, x='Name', y='AverageDose', palette='Greens')
plt.title('Dose media por medicamentos')
plt.xlabel('Bloco')
plt.ylabel('Media de estadia (em dias)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```

```python
query6 = '''
SELECT Department.Name, COUNT(*) AS PhysicianCount
FROM Physician
JOIN Affiliated_With ON Physician.EmployeeID = Affiliated_With.Physician
JOIN Department ON Affiliated_With.Department = Department.DepartmentID
GROUP BY Department.Name;
'''
cursor.execute(query6)
results6 = cursor.fetchall()

df6 = pd.read_sql_query(query6, conn)
print(df6)
plt.figure(figsize=(10, 6))
sns.barplot(data=df6, x='Name', y='PhysicianCount', palette='Greens')
plt.title('Médicos por departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de professionais')
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))  # Display only integers on y-axis
plt.show()
```
